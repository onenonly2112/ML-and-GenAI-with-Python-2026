```python
#PART A
import pandas as pd

# Load Dataset
df = pd.read_excel("agriculture_yield_dataset.csv.xlsx")

# Q1: Dataset Overview

# Number of rows and columns
print("Rows and Columns:")
print(df.shape)

# Names of all columns
print("\nColumn Names:")
print(df.columns.tolist())

# First 10 records
print("\nFirst 10 Records:")
print(df.head(10))


# Q2: Data Types and Missing Values

# Data type of each column
print("\nData Types:")
print(df.dtypes)

# Missing values in each column
print("\nMissing Values:")
print(df.isnull().sum())

# Columns containing missing values
missing_cols = df.columns[df.isnull().sum() > 0]

print("\nColumns with Missing Values:")
if len(missing_cols) > 0:
    print(list(missing_cols))
else:
    print("No missing values found.")


# Q3: Descriptive Statistics

# Summary statistics for numerical columns
stats = df.describe()

print("Summary Statistics:")
print(stats)

# Feature with highest mean value
highest_mean_feature = stats.loc['mean'].idxmax()
print("\nFeature with Highest Mean Value:")
print(highest_mean_feature)
print("Mean =", stats.loc['mean', highest_mean_feature])

# Feature with highest standard deviation
highest_std_feature = stats.loc['std'].idxmax()
print("\nFeature with Highest Standard Deviation:")
print(highest_std_feature)
print("Standard Deviation =", stats.loc['std', highest_std_feature])

#PART B
# Q4: Distribution Analysis

import matplotlib.pyplot as plt

# Histogram for rainfall_mm
plt.figure(figsize=(6,4))
plt.hist(df['rainfall_mm'], bins=20)
plt.title('Distribution of Rainfall')
plt.xlabel('rainfall_mm')
plt.ylabel('Frequency')
plt.show()
# Observations:
# 1. Rainfall values are spread fairly evenly between 300 mm and 1200 mm.
# 2. The distribution appears nearly uniform with no major peaks or gaps.
# 3. No significant outliers are visible in the rainfall data.

# Histogram for temperature_c
plt.figure(figsize=(6,4))
plt.hist(df['temperature_c'], bins=20)
plt.title('Distribution of Temperature')
plt.xlabel('temperature_c')
plt.ylabel('Frequency')
plt.show()
# Observations:
# 1. Temperature values range approximately from 18°C to 38°C.
# 2. The distribution is fairly uniform across the range.
# 3. No extreme values or noticeable outliers are present.

# Histogram for fertilizer_kg
plt.figure(figsize=(6,4))
plt.hist(df['fertilizer_kg'], bins=20)
plt.title('Distribution of Fertilizer')
plt.xlabel('fertilizer_kg')
plt.ylabel('Frequency')
plt.show()
# Observations:
# 1. Fertilizer usage ranges from 50 kg to 250 kg.
# 2. The values are distributed relatively evenly across the range.
# 3. No strong skewness or significant outliers are observed.

# Histogram for yield_ton_per_hectare
plt.figure(figsize=(6,4))
plt.hist(df['yield_ton_per_hectare'], bins=20)
plt.title('Distribution of Yield')
plt.xlabel('yield_ton_per_hectare')
plt.ylabel('Frequency')
plt.show()
# Observations:
# 1. Yield is approximately normally distributed (bell-shaped).
# 2. Most observations are concentrated around 5 tons per hectare.
# 3. Very few values occur at the extremes, indicating no major outliers.

# Q5: Crop Type Analysis

# Number of records for each crop type
crop_counts = df['crop_type'].value_counts()

print("Number of Records for Each Crop Type:")
print(crop_counts)

# Count Plot (Bar Chart)
plt.figure(figsize=(6,4))
crop_counts.plot(kind='bar')
plt.title('Count of Crop Types')
plt.xlabel('Crop Type')
plt.ylabel('Frequency')
plt.show()

# Most frequent crop
print("\nMost Frequent Crop Type:")
print(crop_counts.idxmax())

# Q6: Soil Type Analysis

# Frequency of each soil type
soil_counts = df['soil_type'].value_counts()

print("Frequency of Each Soil Type:")
print(soil_counts)

# Count Plot (Bar Chart)
plt.figure(figsize=(6,4))
soil_counts.plot(kind='bar')
plt.title('Count of Soil Types')
plt.xlabel('Soil Type')
plt.ylabel('Frequency')
plt.show()

# Most common soil type
print("\nMost Common Soil Type:")
print(soil_counts.idxmax())

# Q7: Yield Distribution

plt.figure(figsize=(6,4))
plt.hist(df['yield_ton_per_hectare'], bins=20)
plt.title('Distribution of Yield')
plt.xlabel('yield_ton_per_hectare')
plt.ylabel('Frequency')
plt.show()
# Observations:
# 1. The yield distribution is approximately bell-shaped (normal distribution).
# 2. Most yield values are concentrated around 5 tons per hectare.
# 3. No significant outliers are visible, though a few values occur at the extremes.

# Answers:
# Is the distribution approximately normal? -> Yes
# Are there any noticeable outliers? -> No significant outliers are observed.

# Q8: Scatter Plot Analysis

# Rainfall vs Yield
plt.figure(figsize=(6,4))
plt.scatter(df['rainfall_mm'], df['yield_ton_per_hectare'])
plt.title('Rainfall vs Yield')
plt.xlabel('rainfall_mm')
plt.ylabel('yield_ton_per_hectare')
plt.show()

# Fertilizer vs Yield
plt.figure(figsize=(6,4))
plt.scatter(df['fertilizer_kg'], df['yield_ton_per_hectare'])
plt.title('Fertilizer vs Yield')
plt.xlabel('fertilizer_kg')
plt.ylabel('yield_ton_per_hectare')
plt.show()
# Observations:
# 1. Scatter plots help visualize the relationship between input features and crop yield.

# Q9: Correlation Analysis

import seaborn as sns

# Select only numerical columns
numerical_df = df.select_dtypes(include=['int64', 'float64'])

# Correlation matrix
corr_matrix = numerical_df.corr()

print("Correlation Matrix:")
print(corr_matrix)

# Heatmap
plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()

# Correlation with Yield
yield_corr = corr_matrix['yield_ton_per_hectare'].sort_values(ascending=False)

print("\nCorrelation with Yield:")
print(yield_corr)

print("\nTop 3 Features Most Correlated with Yield:")
print(yield_corr[1:4])  # excluding yield itself
# Observations:
# 1. The heatmap shows the strength and direction of relationships between numerical features.
# 2. Positive values indicate a direct relationship, while negative values indicate an inverse relationship.
# 3. Features with correlation values closest to +1 or -1 have the strongest relationship with yield.

# Q10: Group-Based Analysis

# Average yield for each crop type
crop_avg_yield = df.groupby('crop_type')['yield_ton_per_hectare'].mean()

print("Average Yield by Crop Type:")
print(crop_avg_yield)

# Average yield for each soil type
soil_avg_yield = df.groupby('soil_type')['yield_ton_per_hectare'].mean()

print("\nAverage Yield by Soil Type:")
print(soil_avg_yield)

# Highest average yield crop
print("\nCrop Type with Highest Average Yield:")
print(crop_avg_yield.idxmax())

# Highest average yield soil
print("\nSoil Type with Highest Average Yield:")
print(soil_avg_yield.idxmax())

#Part C

# Q11: Feature Encoding

# Identify categorical columns
categorical_columns = df.select_dtypes(include=['object']).columns

print("Categorical Columns:")
print(list(categorical_columns))

# One-Hot Encoding
df_encoded = pd.get_dummies(df, columns=categorical_columns)

print("\nFirst Five Rows of Transformed Dataset:")
print(df_encoded.head())

# Q12: Feature Selection

# Input Features (X)
X = df_encoded.drop('yield_ton_per_hectare', axis=1)

# Target Variable (y)
y = df_encoded['yield_ton_per_hectare']

print("Input Features (X):")
print(X.head())

print("\nTarget Variable (y):")
print(y.head())

print("\nShape of X:", X.shape)
print("Shape of y:", y.shape)
# Answer:
# Target Variable = yield_ton_per_hectare

#Part D

# Q13: Train-Test Split

from sklearn.model_selection import train_test_split

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.20,
    random_state=42
)

# Display shapes
print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)

# Q14: Linear Regression Model

from sklearn.linear_model import LinearRegression

# Create and train the model
model = LinearRegression()
model.fit(X_train, y_train)

# Display intercept
print("Intercept:")
print(model.intercept_)

# Display coefficients
coefficients = pd.DataFrame({
    'Feature': X.columns,
    'Coefficient': model.coef_
})

print("\nModel Coefficients:")
print(coefficients)

# Feature with highest positive coefficient
highest_positive = coefficients.loc[
    coefficients['Coefficient'].idxmax()
]

print("\nFeature with Highest Positive Coefficient:")
print(highest_positive)
```
